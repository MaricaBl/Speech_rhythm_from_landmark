```python
def stats_from_file(labels_filename:str, time_silence:float=0.2)-> RhythmStatistics:
    try:
        file = leggi_file(labels_filename)
        V_LM, l_intervals = create_list(file)
        interv = create_intervals(l_intervals)
        labels = label_V_or_C(interv, V_LM, time_silence=time_silence)

        return calculate_stats(labels)
    except FileNotFoundError:
        print(f'File {labels_filename} not found')
```

```python
def evaluation_match_from_files(labels_filename: str, annotazioni_filename: str, time_silence=0.2) -> EvaluationMetrics:
    try:
        file = leggi_file(labels_filename)
        V_LM, l_intervals = create_list(file)
        interv = create_intervals(l_intervals)
        labels = label_V_or_C(interv, V_LM, time_silence=time_silence)
        labels = merge_equal_interval(labels)
        
        file_VAL = leggi_file(annotazioni_filename)
        lista_VAL = create_list_VAL(file_VAL)
        labels_VAL = label_intervals_VAL(lista_VAL)
        labels_VAL = merge_equal_interval(labels_VAL)

        return evaluation_match(labels, labels_VAL)
    except FileNotFoundError:
        print(f'File {labels_filename} not found')
```
